```python
from wimprates_ansh import *
import numpy as np
import matplotlib.pyplot as plt
m_MeV = 100
sigmaE = 1e-36

fdm = 0
dside_3s = np.loadtxt('darkside_100MeV_1e-36_fdm0_3s.csv',delimiter=',')
dside_3p = np.loadtxt('darkside_100MeV_1e-36_fdm0_3p.csv',delimiter=',')

drs_100 = dRdlnE(m_MeV,sigmaE,fdm,e_er = None,isoangle = None,mat = 'Ar',useVerne = True,verbose=False,oldParams=True,integrate=False)
# plot_dRdE(drs,m_MeV,sigmaE,e_er=None,mat= 'Ar',lnE = True,plotTotal = False,day=True)
for key in drs_100.keys():
    if key =='3p32':
        continue
    dr = drs_100[key]
    dr = dr* years_per_day
    if key =='3s':
        color = 'red'
    else:
        color = 'blue'
    plt.plot(np.geomspace(1, 400, 100),dr,label=key,color=color)

plt.scatter(dside_3s[:,0],dside_3s[:,1],label='Darkside 3s',color='red',s=0.5)
plt.scatter(dside_3p[:,0],dside_3p[:,1],label='Darkside 3p',color='blue',s=0.5)

plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.xlabel('$E_{er}$ [eV]')
plt.ylabel('dRdlnE (1/kg/day)')
plt.title('100 MeV σ = $10^{-36}$ FDM = 0 Recoil Spectra')
```
